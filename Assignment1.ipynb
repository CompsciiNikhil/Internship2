```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('addiction_population_data.csv')

df.columns = df.columns.str.strip()
df.fillna(0, inplace=True)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='gender', y='smokes_per_day', data=df)
plt.title('Average Cigarettes Smoked Per Day by Gender')
plt.xlabel('Gender')
plt.ylabel('Cigarettes per Day')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['bmi'], bins=20, kde=True)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='mental_health_status', y='attempts_to_quit_smoking', data=df)
plt.title('Quit Smoking Attempts by Mental Health Status')
plt.xlabel('Mental Health Status')
plt.ylabel('Attempts to Quit Smoking')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
